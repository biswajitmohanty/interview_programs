```
import plotly.express as px

gapminder = px.data.gapminder()
gapminder2007 = gapminder.query("year == 2007")
gapminder2007.head()

px.bar(data_frame= gapminder2007,y ='country', x = 'pop',orientation="h")

```

```
fig = px.box(data_frame = gapminder, y= 'lifeExp')
fig.update_traces(boxmean=True)
fig.show()
```

```
fig = px.box(data_frame = gapminder, y= 'lifeExp', color='continent')
fig.update_traces(boxmean=True)
fig.show()
```

```
px.line(gapminder.query("country == ['India','China','United States','Kenya']"),'year','pop', color='country')
```

```
continet_count = gapminder['continent'].value_counts().reset_index()

continet_count.columns = ['continent','count']

px.pie(continet_count, values='count', names='continent', title='Proportion of Continents in Gapminder Dataset')
```

```
df = px.data.carshare()
fig = px.scatter_map(
    data_frame = df,
    lat = 'centroid_lat',
    lon = 'centroid_lon',
    color = 'peak_hour',
    size = 'car_hours',
    color_continuous_scale = px.colors.cyclical.IceFire,
    size_max=15,
    zoom=10,
    map_style='carto-positron'
)
fig.show()
```

```
df = px.data.election()
geojson = px.data.election_geojson()

fig = px.choropleth_map(df,
                        geojson=geojson,
                        color="Bergeron",
                        locations="district",
                        featureidkey="properties.district",
                        center={"lat": 45.5517, "lon": -73.7073},
                        map_style="carto-positron",
                        zoom=9)
fig.show()
```

```
df = px.data.gapminder()
fig = px.choropleth(df,
                    locations="iso_alpha",
                    color="lifeExp",
                    hover_name="country",
                    animation_frame="year",
                    range_color=[20,80])
fig.show()
```
