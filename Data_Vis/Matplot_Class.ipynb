```
# Example 2 - Scatterplot with groups

groups = a[["Promo","Revenue","Marketing Spend"]].groupby("Promo")
print(groups.groups)

fig, ax = plt.subplots(figsize=(15,10))

colors = {
    'Promotion Red': 'red',
    'Promotion Blue': 'blue',
    'No Promo': 'green'
}

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo,  color=colors[promo])


ax.legend()
plt.title("Relationship between M. Spend & Revenue by Promo", fontsize = 20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```
